```python
auto = pd.read_csv('Auto.csv')
auto.head()
```

```python
auto['horsepower'] = pd.to_numeric(auto['horsepower'], errors='coerce')
```

```python
auto.dropna(inplace=True)
```

```python
auto['intercept'] = 1
```

```python
X = auto[['intercept','horsepower', 'mpg', 'displacement', 'weight', 'acceleration']]
y = auto['cylinders']
```

```python
model = sm.OLS(y, X)
results = model.fit()
```

```python
# do the prediction upon intercept = 1 and horsepower = 98
new_predictions = results.get_prediction([1,130,18,300,3500,10]);
new_predictions.predicted_mean
```
