**(e) Use the boxplot() method of college to produce side-by-side boxplots of Outstate versus Private.**

```python
import matplotlib.pyplot as plt
boxplot = college.boxplot(column='Outstate', by='Private')
boxplot.set_title('Outstate tuition by college type')
boxplot.set_xlabel('Private')
boxplot.set_ylabel('Outstate Tuition')
plt.suptitle('')
plt.show()
```
